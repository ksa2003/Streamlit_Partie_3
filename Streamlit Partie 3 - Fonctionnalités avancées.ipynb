```python
import streamlit as st
from streamlit_authenticator import Authenticate
from streamlit_option_menu import option_menu
import pandas as pd

# Charger les donn√©es utilisateur
users_df = pd.read_csv("user_data.csv")
lesDonneesDesComptes = users_df.set_index("name").T.to_dict()

# Configuration de l'authentification
authenticator = Authenticate(
    credentials=lesDonneesDesComptes,
    cookie_name="my_app_cookie",
    key="my_app_key",
    cookie_expiry_days=30
)

# Gestion de l'authentification
name, authentication_status, username = authenticator.login("Login", "main")

if authentication_status:
    authenticator.logout("D√©connexion", "sidebar")
    st.sidebar.write(f"Bienvenue {name} üëã")

    # Menu de navigation
    with st.sidebar:
        selection = option_menu(
            menu_title="Menu principal",
            options=["Accueil", "Photos"],
            icons=["house", "image"]
        )

    # Page d'accueil
    if selection == "Accueil":
        st.title("Bienvenue sur ma page")
        st.image("https://media.giphy.com/media/l0HlSNOxJB956qwfK/giphy.gif")

    # Album photo
    elif selection == "Photos":
        st.title("Bienvenue dans l'album de mon chat üê±")
        col1, col2, col3 = st.columns(3)
        with col1:
            st.image("https://path_to_your_image_1.jpg")
        with col2:
            st.image("https://path_to_your_image_2.jpg")
        with col3:
            st.image("https://path_to_your_image_3.jpg")

elif authentication_status is False:
    st.error("Nom d'utilisateur ou mot de passe incorrect.")
elif authentication_status is None:
    st.warning("Veuillez entrer vos informations.")
```
